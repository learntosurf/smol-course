## Load adapters from the Hub



```python
import torch
import transformers
from datasets import load_dataset
from peft import LoraConfig, get_peft_model
from peft import PeftModel, PeftConfig
from transformers import AutoModelForCausalLM, AutoTokenizer

peft_model_id = "ybelkada/opt-6.7b-lora" # 사전 학습된 LoRA 모델 ID
config = PeftConfig.from_pretrained(peft_model_id) # LoRA 설정 로드
model = AutoModelForCausalLM.from_pretrained( # LoRA 적용 전 기본 모델 로드
    config.base_model_name_or_path,
    return_dict=True,
    load_in_8bit=True, # 8비트 로드로 메모리 절약
    device_map="auto",
)
tokenizer = AutoTokenizer.from_pretrained(config.base_model_name_or_path)

# Load the Lora model
model = PeftModel.from_pretrained(model, peft_model_id)
```

```python
# TO DO: merge with base model
# LoRA 모델 병합
merged_model = model.merge_and_unload()

# 병합된 모델 저장
merged_model.save_pretrained(
    "merged_model", safe_serialization=True, max_shard_size="2GB"
)
tokenizer.save_pretrained("merged_model")
```

```python
del model
torch.cuda.empty_cache()
```

```python
# 병합된 모델 로드 후 Inference 수행
merged_model = AutoModelForCausalLM.from_pretrained(
    "merged_model",
    torch_dtype=torch.float16,
    device_map="auto",
)
tokenizer = AutoTokenizer.from_pretrained("merged_model")
```

```python
batch = tokenizer("Machine learning is: ", return_tensors='pt').to("cuda")
print("Batch input_ids:", batch["input_ids"])

with torch.amp.autocast(device_type="cuda", dtype=torch.float16):
  output_tokens = merged_model.generate(
      **batch,
      max_new_tokens=100,
      do_sample=True,
      temperature=1.0,
      top_p=0.9,
      )

print('\n\n', tokenizer.decode(output_tokens[0], skip_special_tokens=True))
```
